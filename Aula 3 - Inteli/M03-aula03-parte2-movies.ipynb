# TMDB 5000 Movie Dataset

## Importar os dados

```python
import pandas as pd

df = pd.read_csv("tmdb_5000_movies.csv")
df.head()
```

## Idiomas dos 5 primeiros filmes

```python
coluna_idioma = df['original_language']
print(coluna_idioma.head())
```

## Selecionar mais idiomas

```python
idiomas_existentes = df['original_language'].unique().tolist()
print(idiomas_existentes)
```

# Aparições cada idioma

```python
contagem_idiomas = df['original_language'].value_counts()
print(contagem_idiomas)
```

## Paises únicos

```python
coluna_paises_producao = df['production_countries']
paises_unicos = []

for paises_lista in coluna_paises_producao:
	# A função eval() avalia uma expressão em Python.
	# No exemplo abaixo, ela converte uma string como "[{"iso_3166_1": "US", "name": "United States of America"}]" em um objeto Python, representando a lista e o dicionário interno.
	for pais in eval(paises_lista):
		paises_unicos.append(pais['name'])

serie_paises_unicos = pd.Series(paises_unicos).unique()

for pais in serie_paises_unicos:
		print(pais)
```

## Paises e suas produções

```python
coluna_paises_producao = df['production_countries']
contagem_filmes_por_pais = {}

for paises_lista in coluna_paises_producao:
    for pais in eval(paises_lista):
        nome_pais = pais['name']
        contagem_filmes_por_pais[nome_pais] = contagem_filmes_por_pais.get(nome_pais, 0) + 1

for pais, contagem in contagem_filmes_por_pais.items():
    print(f"{pais}: {contagem}")
```

## Produções brasileiras ou produzidos no Brasil

```python
filmes_no_brasil = df[df['production_countries'].str.contains('{"iso_3166_1": "BR", "name": "Brazil"}', na=False)]
print(filmes_no_brasil[['original_title', 'production_countries']])
```

## Gêneros das Produções brasileiras

```python
filmes_brasileiros = df[df['production_countries'].str.contains('{"iso_3166_1": "BR", "name": "Brazil"}', na=False)]
coluna_generos = filmes_brasileiros['genres']
generos_unicos = set()

for generos_lista in coluna_generos:
    for genero in eval(generos_lista):
        generos_unicos.add(genero['name'])

print("Gêneros de Filmes Brasileiros:")
for genero in generos_unicos:
    print(genero)
```

## Overview das Produções brasileiras

```python
filmes_brasileiros = df[df['production_countries'].str.contains('{"iso_3166_1": "BR", "name": "Brazil"}', na=False)]

for index, filme in filmes_brasileiros.iterrows():
    print(f"Título: {filme['original_title']}")
    print(f"Visão Geral: {filme['overview']}")
    print("=" * 50)
```

## Gráfico idiomas encontrados

```python
import matplotlib.pyplot as plt

contagem_idiomas = df['original_language'].value_counts()

plt.figure(figsize=(12, 6))
contagem_idiomas.plot(kind='bar')
plt.title('Contagem de Filmes por Idioma')
plt.xlabel('Idioma')
plt.ylabel('Número de Filmes')
plt.xticks(rotation=45)
plt.show()
```

## Gráfico frequencia de generos

```python
contagem_generos = df['genres'].explode().value_counts()

plt.figure(figsize=(8, 8))
contagem_generos.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Gêneros de Filmes')
plt.show()
```

## Gráfico dispersão para orçamentos e receitas

```python
filmes_orcamento_receita = df[df['budget'] > 0][df['revenue'] > 0]

plt.figure(figsize=(10, 6))
plt.scatter(filmes_orcamento_receita['budget'], filmes_orcamento_receita['revenue'], alpha=0.5)
plt.title('Relação entre Orçamento e Receita de Filmes')
plt.xlabel('Orçamento')
plt.ylabel('Receita')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

## Gráfico evolução das receitas

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Filtrar filmes com data de lançamento não nula e receita positiva
filmes_lancamento_receita = df[df['release_date'].notnull() & (df['revenue'] > 0)]

# Agrupar por ano e calcular a receita total por ano
receita_por_ano = filmes_lancamento_receita.groupby(filmes_lancamento_receita['release_date'].dt.year)['revenue'].sum()

# Criar um gráfico de linhas para visualizar a evolução da receita ao longo dos anos
plt.figure(figsize=(12, 6))
receita_por_ano.plot(kind='line')
plt.title('Evolução da Receita de Filmes ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Receita Total')
plt.xticks(receita_por_ano.index, rotation=45)
plt.show()
```

## Gráfico distribuição avaliações filmes

```python
# Filtrar filmes com avaliações não nulas
filmes_com_avaliacoes = df[df['vote_average'].notnull()]

# Criar um histograma para visualizar a distribuição de avaliações
plt.figure(figsize=(10, 6))
plt.hist(filmes_com_avaliacoes['vote_average'], bins=20, edgecolor='black')
plt.title('Distribuição de Avaliações de Filmes')
plt.xlabel('Avaliação')
plt.ylabel('Frequência')
plt.show()
```

## Gráfico orçamento por gênero

```python
# Filtrar filmes com orçamento não nulo
filmes_com_orcamento = df[df['budget'] > 0]

# Criar um gráfico de boxplot para visualizar a distribuição de orçamentos por gênero
plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)
plt.boxplot([filmes_com_orcamento[filmes_com_orcamento['genres'].str.contains(genero)]['budget'] for genero in generos_unicos],labels=generos_unicos, vert=False)
plt.title('Distribuição de Orçamentos por Gênero')
plt.xlabel('Orçamento')
plt.ylabel('Gênero')
plt.show()
```
